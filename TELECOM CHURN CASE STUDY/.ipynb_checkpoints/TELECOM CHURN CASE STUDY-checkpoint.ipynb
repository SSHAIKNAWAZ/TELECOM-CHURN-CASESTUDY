```python
for col in numeric_cols: 
    q1 = df_new[col].quantile(0.10)
    q3 = df_new[col].quantile(0.90)
    iqr = q3-q1
    range_low  = q1-1.5*iqr
    range_high = q3+1.5*iqr
    # Assigning the filtered dataset into data
    data = df_new.loc[(df_new[col] > range_low) & (df_new[col] < range_high)]

data.shape
```

<font color="#14aaf5 " style ="font-family:Cambria ;font-size:20px"><b> DERIVING A  COLUMN FOR INCOMING AND OUTGOING CALLS MADE IN THE MONTH OF JUNE</b></font><br>

```python
data['total_ic_og_mou_6']  =data['total_og_mou_6']+data['total_ic_mou_6']
```

```python
# we are taking average because we have  twwo months 7 and 8

data['avg_mou_7_8']= (data['total_og_mou_7']+data['total_ic_mou_7']+data['total_ic_mou_8']+data['total_og_mou_8'])/2
```

```python
# Difference avg_mou_good and avg_mou_action
data['diff_mou'] = data['avg_mou_7_8'] - data['total_ic_og_mou_6']
```

```python
# Checking whether the mou has decreased in action phase
data['decrease_mou_action'] = np.where((data['diff_mou'] < 0), 1, 0)
```


<font color="#14aaf5 " style ="font-family:Cambria ;font-size:20px"><b> DERIVING NEW COLUMN DECREASE RECH AMOUNT</b></font><br>
<hr style = "height:1px; color:black">

This column indicates whether the amount of recharge of the customer has decreased in the action phase than the good phase.

```python
data['avg_rech_amount']  =(data['total_rech_amt_7']+data['total_rech_amt_8'])/2
```

```python
data['diff_rech']=data['avg_rech_amount']-data['total_rech_amt_6']
```

```python
# Checking if rech_amt has decreased in action phase
data['decrease_rech_amt_action'] = np.where((data['diff_rech'] < 0), 1, 0) 
```

```python
# VBC in action phase
data['avg_vbc_3g_action'] = (data['jul_vbc_3g'] + data['aug_vbc_3g'])/2
```

```python
# Difference of good and action phase VBC
data['diff_vbc'] = data['avg_vbc_3g_action'] - data['jun_vbc_3g']
```

```python
# Checking whether the VBC has decreased on the action month
data['decrease_vbc_action'] = np.where(data['diff_vbc'] < 0 , 1, 0)
```

<font color="#14aaf5 " style ="font-family:Cambria ;font-size:20px"><b> THESE ARE THE `7 COLUMNS` WHOSE `STANDARD DEVIATION` IS `0` THESE COLUMNS DOESN'T PLAY MAJOR ROLE SO BETTER TO DROP THEM</b></font><br>

```python
def add(true_positive,false_negative,false_positive,true_negative):
    
    c=true_positive / float(true_positive+false_negative)
    print('the senssitivy :' ,c)
    d=true_negative / float(true_negative+false_positive)
    print('the specificity',d)
#     accuraacy
    e=metrics.accuracy_score(y_pred_final.churn_probability,y_pred_final.pridicted)
    print('the accuracy of model is :',e)
#     precison
    f=true_positive/(true_positive+false_positive)
    print('the precision is :',f)
#     recal
    g=true_positive/(true_positive+false_negative)
    print('the recall is ',g)
add(true_positive,false_negative,false_positive,true_negative)
```

<font color="black " style ="font-family:Cambria ;font-size:20px"><b>ROC CURVE </b></font><br>


```python
def draw_roc_curve(actual,probas):
#     code
    fpr,tpr,tresholds=metrics.roc_curve(actual,probas,drop_intermediate=False)
    auc_score=metrics.roc_auc_score(actual,probas)
    plt.plot(fpr,tpr,label='THE ROC CUREVE (AREA=%0.2F)'%auc_score)
    plt.plot((0,1),'k--')
    
#     --------------------------------------------------------------------------------------------
    plt.title('ROC CURVE',fontdict={'size':10,'color':'brown'})
    plt.xlim([0.0,1.2])
    plt.ylim([0.0,1.2])
    plt.xlabel('FALSE POSITIVE RATE',fontdict={'size':7,'color':'red'})
    plt.ylabel('TRUE POSITIVE RATE',fontdict={'size':7,'color':'red'})
    plt.legend(loc='upper left')
    
    
    return
draw_roc_curve(y_pred_final.churn_probability,y_pred_final.churn_prob)
```
